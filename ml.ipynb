## loan_grade
```loan_grade``` = เกรดคุณภาพหนี้ที่ธนาคารจัดลำดับให้ (เช่น A, B, C, D)<br>

````python
df['loan_grade'].unique()
````

- **A**: เป็นเกรดที่ดีที่สุด หมายถึงผู้กู้ที่มีความน่าเชื่อถือสูงมาก มีความเสี่ยงในการผิดนัดชำระหนี้ต่ำ
- **B**: เกรดดีรองลงมา ยังคงมีความน่าเชื่อถือสูง แต่มีความเสี่ยงสูงกว่าเกรด A เล็กน้อย
- **C**: เป็นเกรดปานกลาง มีความน่าเชื่อถือพอสมควร แต่มีความเสี่ยงในการผิดนัดชำระหนี้ในระดับที่ยอมรับได้
- **D**: เกรดที่บ่งชี้ว่ามีความเสี่ยงปานกลางถึงสูงกว่าค่าเฉลี่ย
- **E**: เกรดที่บ่งชี้ว่ามีความเสี่ยงสูง
- **F**: เกรดที่มีความเสี่ยงสูงมาก ผู้กู้อาจมีประวัติทางการเงินที่ไม่ดีนัก
- **G**: เป็นเกรดที่แย่ที่สุด บ่งชี้ถึงผู้กู้ที่มีความเสี่ยงสูงที่สุดในการผิดนัดชำระหนี้

````python
plt.subplots(figsize=(8,5))
sns.countplot(y="loan_grade", data=df, palette='magma')
plt.grid(False)
plt.xlabel('Count')
plt.show()
````
