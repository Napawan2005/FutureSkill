```python
import pandas as pd
df = pd.read_csv('1.txt', delimiter='\t', header=None)
```

**พารามิเตอร์ที่สำคัญ** <br>
`delimiter = '\t' `   : ข้อมูลใน Dataset นี้จะไม่ได้คั่นอยู่รูปแบบ `,` แต่จะถูคั่นในการแบ่ง cloumn คือ `\t` tab แทน

`header = None ` = เมื่ออ่านข้อมูลเข้ามา  head จะกำกับด้วย **index**
