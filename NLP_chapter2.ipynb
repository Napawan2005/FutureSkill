**collections**

- Counter = นับความถี่ของคำ (Word Count)

- defaultdict = การสร้าง Dictionary

```python
model = defaultdict(lambda:defaultdict(lambda:0))
```

- `lambda: 0 (ด้านใน):`  ถ้ามีการเรียก Key ที่ไม่มีอยู่ ให้คืนค่า 0 ทันที

- `defaultdict(...)` : Dictionary ตัวในที่จะถูกสร้างขึ้นมาเก็บค่า 0

- `lambda... (ด้านนอกสุด)` : าเรียก Key หลักแล้วไม่เจอ ให้สร้าง Dictionary ตัวใหม่ขึ้นมา
